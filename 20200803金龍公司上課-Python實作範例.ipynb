```python
df.columns = ['Date', 'Price']
```

```python
df.set_index('Date', drop=True, inplace=True)
```

```python
df.sort_index(inplace=True)
```

```python
def qa_func(question):
  if '保險' in question:
    return '以下是我們推薦的險總類'
  elif '貸款' in question:
    return '以下是我們推薦的貸款總類'
  elif '存款' in question:
    return '以下是您的銀行存款餘額'
  else:
    return '我現在還不聰明, 等我變聰明以後再告訴你'
```

```python
question = input('請輸入你的問題:')
qa_func(question)
```

```python
class Circle(object):
  pi = 3.14
  def __init__(self, radius=1):
    self.radius = radius

  def area(self):
    return self.radius * self.radius * Circle.pi

  def setRadius(self, radius):
    self.radius = radius
    
  def getRadius(self):
    return self.radius
```

```python
a = Circle()
```

```python
a.radius
```

```python
a.getRadius()
```

```python
a.area()
```

```python
a.setRadius(2)
```

```python
a.getRadius()
```

```python
a.area()
```

## File

```python
# r: read, a: append, b:binary
fid = open('test.txt', 'w')
fid.write('Hello\nWorld')
fid.close()
```

```python
with open('test.txt', 'w') as f:
  f.write('Hello\nWorld')
```

```python
f=open('test.txt','r')
for line in f:
    print(line,end="")
f.close()
```

```python
with open('test.txt', 'r') as fid:
  for line in fid:
    print(line,end="")
fid.close()
```

```python
with open('test.txt', 'r') as fid:
  s = fid.read()
```
